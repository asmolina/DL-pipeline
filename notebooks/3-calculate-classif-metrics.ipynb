```python
import numpy as np
```

```python
import rasterio

with rasterio.open('./../data/images/nevelsk/20180423/solar-water-12-2group-pred-stride-32.tif') as src:
    inference = src.read(1) #0: 'лиственные', 1: 'хвойные',
    
with rasterio.open('./../data/images/nevelsk/rasterized/1layer_prevail_group_2groups.tif') as src:
    gt = src.read(1)   #1: 'лиственные', 2: 'хвойные',
```

```python
gt.shape
```

```python
inference.shape
```

```python
np.unique(gt)
```

```python
np.unique(inference)
```

```python
gt -= 1 
np.unique(gt) #0: 'лиственные', 1: 'хвойные',
```

```python
inference_masked = np.where(gt < 255, inference, -1)
gt_masked = np.where(gt < 255, gt, -1)
```

```python
import tifffile as tiff
```

```python
tiff.imshow(inference_masked)
```

```python
tiff.imshow(gt_masked)
```

```python
np.unique(gt_masked), np.unique(inference_masked)
```

```python
from sklearn.metrics import classification_report
```

```python
Y_gt = gt_masked[gt_masked > -1]
Y_inference = inference_masked[gt_masked > -1]
```

```python
np.unique(Y_gt),  np.unique(Y_inference)
```

```python
Y_gt.shape, Y_inference.shape
```

```python
from sklearn.metrics import recall_score, f1_score, precision_score

recall_score(Y_gt, Y_inference, average=None, labels = ['лиственные', 'хвойные'])
```

```python
print(classification_report(Y_gt, Y_inference))
#  labels = ['лиственные', 'хвойные'] не будет
```
